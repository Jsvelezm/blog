```
# hide
datos_url = 'https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_19-covid-Confirmed.csv&filename=time_series_2019-ncov-Confirmed.csv'
datos = pd.read_csv(datos_url)
datos_mundiales  = datos.groupby(['Country/Region']).sum()
```

```
# hide
latinos =[ 'Argentina','Bolivia','Brazil','Chile','Colombia','Costa Rica','Cuba',
          'Ecuador','El Salvador','Granada','Guatemala','Guayana','Guayana Francesa',
          'Haiti','Honduras','Jamaica','Mexico','Nicaragua','Panama','Paraguay',
          'Peru','Puerto Rico','Dominica','Suriname','Uruguay'
          ,'Venezuela']
paises = datos['Country/Region']
datos_latinos = datos[[x in latinos for x in paises]] 
```

```
# hide
date_time= datetime.datetime.now()
date_time = str(date_time)
"".join(date_time.split())
datos_mundiales.to_excel('covid-mundial'+ date_time +'.xlsx')
```

```
# hide

folium_map = folium.Map(location=[0, 0],
                        zoom_start = 2.5,
                        tiles="CartoDB dark_matter")

```

```
# hide
total = datos.iloc[:,4:].sum().sum()
for index, row in datos.iterrows():
    
    casos = (row['3/9/20'])
    radius = casos/total
    #if net_departures>0:
    color="#E37222" # tangerine
    #else:
    #color="#0A8A9F" # teal
    
    folium.CircleMarker(location=(row["Lat"],
                                  row["Long"]),
                        radius=radius,
                        color=color,
                        fill=True).add_to(folium_map)
```
